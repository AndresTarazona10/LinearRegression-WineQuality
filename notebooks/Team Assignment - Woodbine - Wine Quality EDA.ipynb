```python
### Load relevant packages
import pandas                  as pd
import numpy                   as np
import matplotlib.pyplot       as plt
import seaborn                 as sns
```

```python
wine_quality = pd.read_csv("../data/winequality-white.csv",sep=';')
# Step for converting white space in columns to _ value for better handling 
wine_quality.rename(columns=lambda x: x.replace(" ", "_"), inplace=True)
```

```python
wine_quality
```

```python
wine_quality.describe()
```

We can see above that the table contains `quality` as dependent variable and 11 independent variables. The quality can aquire an integer value from 3 to 9, with a median value of 6. In the following graph we can observe the lower and higer quality have less representation in the dataset.

```python
# Print a bar chart of the quality column
wine_quality['quality'].value_counts().sort_index().plot(kind='bar')
plt.title('Wine quality')
plt.xlabel('Category')
plt.ylabel('Count')
```

Let's take a look at the independent variables

```python
fig, axes = plt.subplots(2, 3,figsize=(12,8))

plt.subplot(2,3,1)
plt.hist(wine_quality.fixed_acidity,bins=20)
plt.xlabel("fixed_acidity")
plt.ylabel("frequency")

plt.subplot(2,3,2)
plt.hist(wine_quality.volatile_acidity,bins=20)
plt.xlabel("fixed_acidity")
plt.ylabel("frequency")

plt.subplot(2,3,3)
plt.hist(wine_quality.citric_acid,bins=20)
plt.xlabel("citric_acid")
plt.ylabel("frequency")

# Print corrected histograms
plt.subplot(2,3,4)
plt.hist(np.log(wine_quality.fixed_acidity),bins=20)
plt.xlabel("fixed_acidity corrected")
plt.ylabel("frequency")

plt.subplot(2,3,5)
plt.hist(np.cbrt(wine_quality.volatile_acidity),bins=20)
plt.xlabel("fixed_acidity corrected")
plt.ylabel("frequency")

plt.subplot(2,3,6)
plt.hist(np.sqrt(wine_quality.citric_acid),bins=20)
plt.xlabel("citric_acid corrected")
plt.ylabel("frequency")

plt.show()
```

```python
fig, axes = plt.subplots(2, 3,figsize=(12,8))

plt.subplot(2,3,1)
plt.hist(wine_quality.residual_sugar,bins=20)
plt.xlabel("residual_sugar")
plt.ylabel("frequency")

plt.subplot(2,3,2)
plt.hist(wine_quality.chlorides,bins=20)
plt.xlabel("chlorides")
plt.ylabel("frequency")

plt.subplot(2,3,3)
plt.hist(wine_quality.free_sulfur_dioxide,bins=20)
plt.xlabel("free_sulfur_dioxide")
plt.ylabel("frequency")

# Print corrected histograms
plt.subplot(2,3,4)
plt.hist(np.cbrt(wine_quality.residual_sugar),bins=20)
plt.xlabel("residual_sugar corrected")
plt.ylabel("frequency")

plt.subplot(2,3,5)
plt.hist(1/np.cbrt(wine_quality.chlorides),bins=20)
plt.xlabel("chlorides corrected")
plt.ylabel("frequency")

plt.subplot(2,3,6)
plt.hist(np.log(wine_quality.free_sulfur_dioxide),bins=20)
plt.xlabel("free_sulfur_dioxide corrected")
plt.ylabel("frequency")

plt.show()
```

```python
fig, axes = plt.subplots(2, 3,figsize=(12,8))

plt.subplot(2,3,1)
plt.hist(wine_quality.total_sulfur_dioxide,bins=20)
plt.xlabel("total_sulfur_dioxide")
plt.ylabel("frequency")

plt.subplot(2,3,2)
plt.hist(wine_quality.density,bins=20)
plt.xlabel("density")
plt.ylabel("frequency")

plt.subplot(2,3,3)
plt.hist(wine_quality.pH,bins=20)
plt.xlabel("pH")
plt.ylabel("frequency")

# Print corrected subplots
plt.subplot(2,3,4)
plt.hist(np.sqrt(wine_quality.total_sulfur_dioxide),bins=20)
plt.xlabel("total_sulfur_dioxide")
plt.ylabel("frequency")

plt.subplot(2,3,5)
plt.hist(1/(wine_quality.density ** 3),bins=20)
plt.xlabel("density corrected")
plt.ylabel("frequency")

plt.subplot(2,3,6)
plt.hist(1/np.sqrt(wine_quality.pH),bins=20)
plt.xlabel("pH corrected")
plt.ylabel("frequency")


plt.show()
```

```python
fig, axes = plt.subplots(2, 2,figsize=(8,8))

plt.subplot(2,2,1)
plt.hist(wine_quality.sulphates,bins=20)
plt.xlabel("sulphates")
plt.ylabel("frequency")

plt.subplot(2,2,2)
plt.hist(wine_quality.alcohol,bins=20)
plt.xlabel("alcohol")
plt.ylabel("frequency")

# Print corrected histograms
plt.subplot(2,2,3)
plt.hist(1/np.cbrt(wine_quality.sulphates),bins=20)
plt.xlabel("sulphates corrected")
plt.ylabel("frequency")

plt.subplot(2,2,4)
plt.hist(1/(wine_quality.alcohol),bins=20)
plt.xlabel("alcohol corrected")
plt.ylabel("frequency")

plt.show()
```

Explore correlation between independend variables

```python
colnms = ['fixed_acidity', 'volatile_acidity', 'citric_acid', 'residual_sugar',
 'chlorides', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'density',
 'pH', 'sulphates', 'alcohol']

# Correlation coefficients
corr_mat = np.corrcoef(wine_quality[colnms].values.T)
sns.set(font_scale=1)
sns.set(rc={"figure.figsize":(11, 11)})
full_mat = sns.heatmap(corr_mat, cbar=True, annot=True, square=True,
fmt='.2f',annot_kws={'size': 15}, yticklabels=colnms,
xticklabels=colnms)
plt.show()
```

## Conclusions

`fixed_acidity`: It presents an slight skewness to the right, this can be corrected by applying a logarithmic transformation to the values.

`volatile_acidity`: It presents left skewness, this can be corrected by applying cubic root transformation to the values.

`citric_acid`: It presents a light left skewness, this can be corrected by applying cubic root transformation to the values.

`residual_sugar`: It is heavely left skewed, the best correction was achieved by aplying square root transformation to the values.

`chlorides`: It is left skewed, the best correction was achieved by aplying the inverse cubic root transforation to the values.

`free_sulfur_dioxide`: If is left squewed, logarithmic transformation shown an improvement in the data distribution.

`total_sulfur_dioxide`: It is lightly skewed to the left, square root transformation shown an improvement in the data distribution.

`density`: It presents negative skewness that is not possible to correct. However it is highly correlated to the `alcohol` (-0.78) and `residual_sugar` (0.84) level, it can be considered to remove this column from the estiamtors, based on the premise it won't affect other features.

`pH`: It was lightly skewed to the left, the inverse square root transformation shown an improvement in the distribution of the data.

`sulphates`: It is skewed to the left, the inverse cubic root enhanced the data distribution.

`alcohol`: It exhibits left skewness, and the inverse transformation has enhanced the distribution of the data.

The left skewness seems to be a predominant for all independent variables, it is recommended to use transformed variables for the model building.
